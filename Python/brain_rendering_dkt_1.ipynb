```python
import neuromaps
import abagen
dkt = abagen.datasets.fetch_desikan_killiany(surface=True)
dkt_image = dkt['image']
dkt_info = pd.read_csv(dkt['info'])
print(dkt_info)

dkt_lh = neuromaps.images.load_gifti(dkt_image[0])
dkt_rh = neuromaps.images.load_gifti(dkt_image[1])
parcellation = (dkt_lh, dkt_rh)
```

```python
pib1_z = pd.concat([dkt_info, pib1_zscores], axis=1)
pib1_z = pib1_z[pib1_z['structure'] == 'cortex']
pib1_z['pib1_mean']=pib1_z.iloc[:34, 4:].mean(axis=1, skipna=True)
pib1_mean = pib1_z['pib1_mean'].to_numpy()

pib2_z = pd.concat([dkt_info, pib2_zscores], axis=1)
pib2_z = pib2_z[pib2_z['structure'] == 'cortex']
pib2_z['pib2_mean']=pib2_z.iloc[:34, 4:].mean(axis=1, skipna=True)
pib2_mean = pib2_z['pib2_mean'].to_numpy()

pib3_z = pd.concat([dkt_info, pib3_zscores], axis=1)
pib3_z = pib3_z[pib3_z['structure'] == 'cortex']
pib3_z['pib3_mean']=pib3_z.iloc[:34, 4:].mean(axis=1, skipna=True)
pib3_mean = pib3_z['pib3_mean'].to_numpy()

pib4_z = pd.concat([dkt_info, pib4_zscores], axis=1)
pib4_z = pib4_z[pib4_z['structure'] == 'cortex']
pib4_z['pib4_mean']=pib4_z.iloc[:34, 4:].mean(axis=1, skipna=True)
pib4_mean = pib4_z['pib4_mean'].to_numpy()

tau1_z = pd.concat([dkt_info, tau1_zscores], axis=1)
tau1_z = tau1_z[tau1_z['structure'] == 'cortex']
tau1_z['tau1_mean']=tau1_z.iloc[:34, 4:].mean(axis=1, skipna=True)
tau1_mean = tau1_z['tau1_mean'].to_numpy()

tau2_z = pd.concat([dkt_info, tau2_zscores], axis=1)
tau2_z = tau2_z[tau2_z['structure'] == 'cortex']
tau2_z['tau2_mean']=tau2_z.iloc[:34, 4:].mean(axis=1, skipna=True)
tau2_mean = tau2_z['tau2_mean'].to_numpy()

tau3_z = pd.concat([dkt_info, tau3_zscores], axis=1)
tau3_z = tau3_z[tau3_z['structure'] == 'cortex']
tau3_z['tau3_mean']=tau3_z.iloc[:34, 4:].mean(axis=1, skipna=True)
tau3_mean = tau3_z['tau3_mean'].to_numpy()

tau4_z = pd.concat([dkt_info, tau4_zscores], axis=1)
tau4_z = tau4_z[tau4_z['structure'] == 'cortex']
tau4_z['tau4_mean']=tau4_z.iloc[:34, 4:].mean(axis=1, skipna=True)
tau4_mean = tau4_z['tau4_mean'].to_numpy()
```
